```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (7, 7))
ax = fig.add_subplot(projection='3d')

X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 = 2*X + 3*Y
Z2 = 2*X + 3*Y + 10
ax.plot_surface(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2)
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (7, 7))
ax = fig.add_subplot(projection='3d')

a = 7
b = 2
c = 5
x = np.linspace(-a, a, 51)
y = np.linspace(-b, b, 51)
x, y = np.meshgrid(x, y)
z = c**2 * (x**2 / (2 * a**2) + y**2 / (2 * b**2))
ax.plot_wireframe(x, y, z)
show()
```
