```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x0 = 2
y0 = 10
R = 5
x = np.linspace(x0 - R, x0 + R, 201)
y_circle_1 = []
y_circle_2 = []
for i in range(len(x)):
    y_circle_1.append(math.sqrt(R**2 - (x[i] - x0)**2) + y0)
    y_circle_2.append(-math.sqrt(R**2 - (x[i] - x0)**2) + y0)
plt.plot(x, y_circle_1, 'cyan')
plt.plot(x, y_circle_2, 'blue')
plt.scatter(x0, y0)
plt.axis('equal')
```

```python
x0 = 0
y0 = 10
a = 4
b = 2
x = np.linspace(x0 - a, x0 + a, 201)
y_ellipse_1 = []
y_ellipse_2 = []
for i in range(len(x)):
    y_ellipse_1.append(math.sqrt(b**2 * (1 - (x[i] - x0)**2 / a**2)) + y0)
    y_ellipse_2.append(-math.sqrt(b**2 * (1 - (x[i] - x0)**2 / a**2)) + y0)
plt.plot(x, y_ellipse_1, 'orange')
plt.plot(x, y_ellipse_2, 'orange')
plt.scatter(x0, y0)
plt.axis('equal')
```

```python
x0 = 0
y0 = 10
a = 5
b = 2
x1 = np.linspace(x0 - 2 * a, x0 - a, 101)
x2 = np.linspace(x0 + a, x0 + 2 * a, 101)
x = list(x1) + list(x2)
y_hyperbola_1 = []
y_hyperbola_2 = []
for i in range(len(x)):
    y_hyperbola_1.append(-math.sqrt(b**2 * ((x[i] - x0)**2 / a**2 - 1)) + y0)
    y_hyperbola_2.append(math.sqrt(b**2 * ((x[i] - x0)**2 / a**2 - 1)) + y0)
plt.scatter(x, y_hyperbola_1)
plt.scatter(x, y_hyperbola_2)
plt.scatter(x0, y0)
plt.axis('equal')
```

```python
x0 = 0
y0 = 0
a = 8
b = 5
x1 = np.linspace(x0 - 2 * a, x0 - a, 101)
x2 = np.linspace(x0 + a, x0 + 2 * a, 101)
y_hyperbola_11 = []
y_hyperbola_12 = []
y_hyperbola_21 = []
y_hyperbola_22 = []
for i in range(len(x1)):
    y_hyperbola_11.append(-math.sqrt(b**2 * ((x1[i] - x0)**2 / a**2 - 1)) + y0)
    y_hyperbola_21.append(math.sqrt(b**2 * ((x1[i] - x0)**2 / a**2 - 1)) + y0)
for i in range(len(x2)):
    y_hyperbola_12.append(-math.sqrt(b**2 * ((x2[i] - x0)**2 / a**2 - 1)) + y0)
    y_hyperbola_22.append(math.sqrt(b**2 * ((x2[i] - x0)**2 / a**2 - 1)) + y0)
plt.plot(x1, y_hyperbola_11, 'green')
plt.plot(x1, y_hyperbola_21, 'green')
plt.plot(x2, y_hyperbola_12, 'green')
plt.plot(x2, y_hyperbola_22, 'green')
plt.scatter(x0, y0)
plt.axis('equal')
```
