```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def polar_change(x, y):
    alpha = []
    R = []
    for i in range(len(x)):
        r = math.sqrt(x[i]**2 + y[i]**2)
        al = np.arctan2(y[i], x[i])
        alpha.append(al)
        R.append(r)
    return alpha, R
x0 = [1, 2]
y0 = [-3, -4]
polar_change(x0, y0)
```

```python
r = 5
x = np.linspace(-r, r, 101)
y_circle_1 = []
y_circle_2 = []
for i in range(len(x)):
    y_circle_1.append(math.sqrt(r**2 - x[i]**2))
    y_circle_2.append(-math.sqrt(r**2 - x[i]**2))
alpha_1 = polar_change(x, y_circle_1)[0]
R1 = polar_change(x, y_circle_1)[1]
alpha_2 = polar_change(x, y_circle_2)[0]
R2 = polar_change(x, y_circle_2)[1]
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(alpha_1, R1)
ax.plot(alpha_2, R2)
```

```python
x = np.linspace(-5, 5, 101)
k = 3
b = 1
y = k * x + b
alpha = polar_change(x, y)[0]
R = polar_change(x, y)[1]
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(alpha, R)
```
